# **Pre-processing - DATASET 02**

```python
IMAGE_SIZE = 32
image_transforms = transforms.Compose([
    transforms.Resize((IMAGE_SIZE,IMAGE_SIZE)),
    transforms.CenterCrop(IMAGE_SIZE), 
    torchvision.transforms.ColorJitter(brightness=0.4,saturation=0.4,contrast=0.4,hue=0.4),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])
```

```python
image_data = torchvision.datasets.ImageFolder(
                              "/content/drive/MyDrive/imagery", 
                              transform = image_transforms)

image_index = np.arange(len(image_data))
image_lebel = image_data.targets

train_x, test_x, train_y, test_y = model_selection.train_test_split(
                                                  image_index, image_lebel, 
                                                  test_size=0.8, stratify=image_lebel, 
                                                  shuffle=True)
```

```python
test_image_datasets = tdata.Subset(image_data, test_x)
train_image_datasets = tdata.Subset(image_data, test_x)

test_loader = DataLoader(test_image_datasets, batch_size = 20)
train_loader = DataLoader(train_image_datasets, batch_size = 20)
```

# **NN**

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(32*32*3, 100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, 4),
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return F.log_softmax(logits, dim=1)
      
NNmodel = NeuralNetwork()
NNmodel

optimizer = optim.Adam(NNmodel.parameters(), lr=0.001)
criterion = torch.nn.CrossEntropyLoss() 
```

```python
EPOCHS = 20
Avg_loss = 0
epoch_loss =[]

for epoch in range(EPOCHS):
    
    for data in train_loader:
        image_data, image_lebel = data
        optimizer.zero_grad()
        logs = NNmodel(image_data)
        loss = criterion(logs, image_lebel)
        loss.backward()
        optimizer.step()
        
    epoch_loss.append(loss.item())
    Avg_loss += loss.item()
    print('epoch: {}. Loss: {}'.format(epoch+1, loss.item()))
    
print('\nAverage Loss: ',Avg_loss/EPOCHS)
```

```python
total1 = 0
total2 = 0
correct1 = 0
correct2 = 0

with torch.no_grad():
    NNmodel.eval()

    for data in test_loader:
        data_input, labels = data
        outputs = NNmodel(data_input)
        _, predicted = torch.max(outputs.data, 1)
        total1 += labels.size(0)
        correct1 += (predicted == labels).sum().item()

    for data in train_loader:
        data_input, labels = data
        outputs = NNmodel(data_input)
        _, predicted = torch.max(outputs.data, 1)
        total2 += labels.size(0)
        correct2 += (predicted == labels).sum().item()

print('Accuracy of the network on the test images: %d %%' % (
    100 * correct1 / total1))
print('Accuracy of the network on the train images: %d %%' % (
    100 * correct2 / total2))

plt.plot(epoch_loss)
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.show()
print('\nAverage Loss: ',Avg_loss/EPOCHS)
```

# **CNN**

```python
class CNeuralNetwork(nn.Module):
    def __init__(self):
        super(CNeuralNetwork, self).__init__()
        self.flatten = nn.Flatten()

        self.conv1 = nn.Conv2d(in_channels=3, out_channels=256,kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(in_channels=256, out_channels=128,kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(in_channels=128, out_channels=64,kernel_size=3, stride=1, padding=1)
        self.conv4 = nn.Conv2d(in_channels=64, out_channels=32,kernel_size=3, stride=1, padding=1)

        """"
        self.fc1 = nn.Linear(128, 256)
        self.fc2 = nn.Linear(256, 128) 
        self.fc3 = nn.Linear(128, 64) 
        self.fc4 = nn.Linear(64, 32)
        self.fc5 = nn.Linear(32, 4)
        """""
        self.fc1 = nn.Linear(128, 512)
        self.fc2 = nn.Linear(512, 256) 
        self.fc3 = nn.Linear(256, 128) 
        self.fc4 = nn.Linear(128, 64)
        self.fc5 = nn.Linear(64, 4)
        
    def forward(self, x):
        x = F.dropout(self.conv1(x), .2)
        x = F.max_pool2d(F.relu(x),2)
        
        x = F.dropout(self.conv2(x), .2)
        x = F.max_pool2d(F.relu(x),2)
        
        x = F.dropout(self.conv3(x), .2)
        x = F.max_pool2d(F.relu(x),2)
        
        x = F.dropout(self.conv4(x), .2)
        x = F.max_pool2d(F.relu(x),2)
       
        x = self.flatten(x)
        
        x = F.relu(self.fc1(x)) # layer 1
        x = F.relu(self.fc2(x)) # layer 2
        x = F.relu(self.fc3(x)) # layer 3
        x = F.relu(self.fc4(x)) # layer 4
        x = F.dropout(x,.2)
        logits = self.fc5(x)    # layer 5
       
        return F.log_softmax(logits, dim=1)

CNNmodel = CNeuralNetwork()
CNNmodel

optimizer = optim.Adam(CNNmodel.parameters(), lr=0.001)
criterion = torch.nn.CrossEntropyLoss() 
```

```python
EPOCHS = 20
Avg_loss = 0
epoch_loss =[]

for epoch in range(EPOCHS):
    
    for data in train_loader:
        image_data, image_lebel = data
        optimizer.zero_grad()
        logs = CNNmodel(image_data)
        loss = criterion(logs, image_lebel)
        loss.backward()
        optimizer.step()
        
    epoch_loss.append(loss.item())
    Avg_loss += loss.item()
    print('epoch: {}. Loss: {}'.format(epoch+1, loss.item()))
    
print('\nAverage Loss: ',Avg_loss/EPOCHS)
```

```python
total1 = 0
total2 = 0
correct1 = 0
correct2 = 0

with torch.no_grad():
    CNNmodel.eval()

    for data in test_loader:
        data_input, labels = data
        outputs = CNNmodel(data_input)
        _, predicted = torch.max(outputs.data, 1)
        total1 += labels.size(0)
        correct1 += (predicted == labels).sum().item()

    for data in train_loader:
        data_input, labels = data
        outputs = CNNmodel(data_input)
        _, predicted = torch.max(outputs.data, 1)
        total2 += labels.size(0)
        correct2 += (predicted == labels).sum().item()

print('Accuracy of the network on the test images: %d %%' % (
    100 * correct1 / total1))
print('Accuracy of the network on the train images: %d %%' % (
    100 * correct2 / total2))

plt.plot(epoch_loss)
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.show()
print('\nAverage Loss: ',Avg_loss/EPOCHS)
```
